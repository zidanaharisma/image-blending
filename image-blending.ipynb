```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

lego_a = cv2.imread('../input/fotobaru/photo_2022-07-04_08-52-16.jpg',1)
lego_a = cv2.cvtColor(lego_a, cv2.COLOR_BGR2RGB)

lego_b = cv2.imread('../input/fotoke2/photo_2022-07-04_09-20-22.jpg',1)
lego_b = cv2.cvtColor(lego_b, cv2.COLOR_BGR2RGB)

lego_blending = cv2.addWeighted(lego_a, 0.7, lego_b, 0.5,0)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_blending)
plt.show()
```
